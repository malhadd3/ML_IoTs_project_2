```python
import os
from pydub import AudioSegment
```

```python
files = os.listdir(os.getcwd())
output_path = 'converted\\'
```

```python
for i, file in enumerate(files):
    list_ban = ['.ipynb_checkpoints','audio_converter.ipynb','converted']
    if file not in list_ban:
        cmd = f"ffmpeg -i {file} -acodec pcm_s16le -ac 1 -ar 16000 {output_path}Active{i}.wav"
        print(cmd)
        os.system(cmd)
        sound = AudioSegment.from_file(f"{output_path}Active{i}.wav")
        sound = sound[:1000]
        sound.export(f"{output_path}Active{i}.wav", format="wav")
        print("-----")
```
